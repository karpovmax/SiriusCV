```python
print(imgBorder.shape)
print('----')
colorBorder = imgBorder[0,0]
for i in range(647):
    flag = 1
    for j in range(826):
        if imgBorder[i, j].all() != colorBorder.all():
            flag = 0
            print(i, j)
            break
    if flag == 0:
        print('up', end='\n----\n')
        break

for i in range(826, 0, -1):
    flag = 1
    for j in range(647):
        if imgBorder[j, i].all() != colorBorder.all():
            flag = 0
            print(j, i)
            break
    if flag == 0:
        print('end')
        break
            
```
